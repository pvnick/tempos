##### interpolate the pain score values across consistent time intervals so that our SAX algorithm aligns accurately with real time

```python
def interpolate_patient_pain_scores(patient_dataframe_multiindexed, interval_minutes):
    #pandas automatically re-adds this in post-processing. remove it so we dont have a duplicate
    patient_dataframe_multiindexed.index = patient_dataframe_multiindexed.index.droplevel(0)
    patient_dataframe_multiindexed.index
    #delete duplicate index items
    patient_dataframe_multiindexed["NRSTimeFromEndSurgery_mins"] = patient_dataframe_multiindexed.index
    patient_dataframe_multiindexed.drop_duplicates(cols='NRSTimeFromEndSurgery_mins', take_last=True, inplace=True)
    patient_dataframe_multiindexed.index = patient_dataframe_multiindexed["NRSTimeFromEndSurgery_mins"]
    del patient_dataframe_multiindexed["NRSTimeFromEndSurgery_mins"]
    
    if patient_dataframe_multiindexed.PainScoreQuantMissing.count() <= 1:
        return patient_dataframe_multiindexed
    patient_dataframe_multiindexed.dropna(subset=["PainScoreQuantMissing"], inplace=True)
    patient_dataframe_unindexed = patient_dataframe_multiindexed.reset_index()
    patient_dataframe_time_indexed = patient_dataframe_unindexed.set_index("NRSTimeFromEndSurgery_mins").sort_index()
    #use linear interpolation from scipy
    interpolation_func = interp1d(x=patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins, 
                                  y=patient_dataframe_unindexed.PainScoreQuantMissing)
    first_time_value = patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins.iloc[0]
    last_time_value = patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins.iloc[-1]
    constant_time_interval_points = np.arange(first_time_value, last_time_value, interval_minutes)
    #expand the original data structure and add the interpolated values
    total_indexed_by_time = patient_dataframe_unindexed.set_index("NRSTimeFromEndSurgery_mins")
    total_indexed_by_time_expanded = total_indexed_by_time.reindex(constant_time_interval_points, method='ffill')
    total_indexed_by_time_expanded["PainScoreQuantMissing"] = [interpolated_val if interpolated_val > 0 else 0 for interpolated_val in interpolation_func(constant_time_interval_points)]
    return total_indexed_by_time_expanded

pain_data_indexed = start_cutoff_pain_data.set_index(["PatientID", "NRSTimeFromEndSurgery_mins"])
grouped_by_patient = pain_data_indexed.groupby(level=0)
pain_interpolated = grouped_by_patient.apply(interpolate_patient_pain_scores, 10)
```

## SAX Calculation

```python
def z_normalize(series):
    if (series.count() == 1):
        return pd.Series([1])
    else:
        std_dev = series.std()
        mean = series.mean()
        ret_series = (series - mean) / std_dev 
        ret_series.fillna(value = 0, inplace = True)
        return ret_series
```

```python
def piecewise_aggregate_approximation(normalized_series, paa_size):
    paa_size = int(paa_size)
    series_length = int(normalized_series.count())
    #take care of the special case where there is no dimensionality reduction
    if series_length == paa_size:
        paa = normalized_series
    else:
        if series_length % paa_size == 0:
            paa = normalized_series.reshape([series_length / paa_size, paa_size], order="F").mean(axis=0)
        else:
            temp = pd.DataFrame(normalized_series).T.reindex(index = xrange(paa_size), method="ffill")
            expanded_sub_section = temp.values.reshape(paa_size * series_length, order="F")
            paa = expanded_sub_section.reshape([series_length, paa_size], order="F").mean(axis=0)
    return pd.Series(paa)
```

```python
def get_cuts(paa_series, alphabet_size):
    alphabet_size = int(alphabet_size)
    cut_points = {2: [-np.inf, 0],
                  3: [-np.inf, -0.43, 0.43],
                  4: [-np.inf, -0.67, 0, 0.67],
                  5: [-np.inf, -0.84, -0.25, 0.25, 0.84],
                  6: [-np.inf, -0.97, -0.43, 0, 0.43, 0.97],
                  7: [-np.inf, -1.07, -0.57, -0.18, 0.18, 0.57, 1.07],
                  8: [-np.inf, -1.15, -0.67, -0.32, 0, 0.32, 0.67, 1.15],
                  9: [-np.inf, -1.22, -0.76, -0.43, -0.14, 0.14, 0.43, 0.76, 1.22],
                  10: [-np.inf, -1.28, -0.84, -0.52, -0.25, 0., 0.25, 0.52, 0.84, 1.28],
                  11: [-np.inf, -1.34, -0.91, -0.6, -0.35, -0.11, 0.11, 0.35, 0.6, 0.91, 1.34],
                  12: [-np.inf, -1.38, -0.97, -0.67, -0.43, -0.21, 0, 0.21, 0.43, 0.67, 0.97, 1.38],
                  13: [-np.inf, -1.43, -1.02, -0.74, -0.5, -0.29, -0.1, 0.1, 0.29, 0.5, 0.74, 1.02, 1.43],
                  14: [-np.inf, -1.47, -1.07, -0.79, -0.57, -0.37, -0.18, 0, 0.18, 0.37, 0.57, 0.79, 1.07, 1.47],
                  15: [-np.inf, -1.5, -1.11, -0.84, -0.62, -0.43, -0.25, -0.08, 0.08, 0.25, 0.43, 0.62, 0.84, 1.11, 1.5],
                  16: [-np.inf, -1.53, -1.15, -0.89, -0.67, -0.49, -0.32, -0.16, 0, 0.16, 0.32, 0.49, 0.67, 0.89, 1.15, 1.53],
                  17: [-np.inf, -1.56, -1.19, -0.93, -0.72, -0.54, -0.38, -0.22, -0.07, 0.07, 0.22, 0.38, 0.54, 0.72, 0.93, 1.19, 1.56],
                  18: [-np.inf, -1.59, -1.22, -0.97, -0.76, -0.59, -0.43, -0.28, -0.14, 0, 0.14, 0.28, 0.43, 0.59, 0.76, 0.97, 1.22, 1.59],
                  19: [-np.inf, -1.62, -1.25, -1, -0.8, -0.63, -0.48, -0.34, -0.2, -0.07, 0.07, 0.2, 0.34, 0.48, 0.63, 0.8, 1, 1.25, 1.62],
                  20: [-np.inf, -1.64, -1.28, -1.04, -0.84, -0.67, -0.52, -0.39, -0.25, -0.13, 0, 0.13, 0.25, 0.39, 0.52, 0.67, 0.84, 1.04, 1.28, 1.64]}
    if alphabet_size not in cut_points:
        raise Exception("Invalid alphabet size")
    relevant_cut_points = cut_points[alphabet_size]
    relevant_cut_points_series = pd.Series(relevant_cut_points)
    cuts = paa_series.apply(lambda paa_element: np.sum(relevant_cut_points <= paa_element))
    return cuts
```

```python
def sax_unwindowed(series, alphabet_size, length_override = None):
    #the following is copied/translated from matlab/symbolic_visual.m (Keogh et al)
    n = 16
    length = series.count()
    # normalize the time series first
    normalized = z_normalize(series)
    # if we represent each symbol as a binary string, calculate how much space 
    # (i.e. how many bits) is needed to store one symbol
    num_bits_per_symbol = math.floor(math.log(alphabet_size, 2)+1)
    #64 bits vs. 3 bits (for alphabet size between 4 and 7) per PAA coef
    PAA_symbolic_ratio = math.floor(64 / num_bits_per_symbol)          
    # this is the maximum symbolic segments possible, using the same amount of space as PAA
    # (assuming each PAA segment uses 8 bytes)
    temp1 = PAA_symbolic_ratio * n                        
    if temp1 >= length:
        sym_seg = length
    # can't just use the maximum # of segments possible -- it has to be divisible by the 
    # original length
    else:
        temp2 = math.floor(math.log(temp1, 2));    
        sym_seg = 2 ** temp2;

    if sym_seg > 4 * n:
        sym_seg = sym_seg / 4;
        
    #overwrite the calculated length with our own length
    if length_override:
        sym_seg = length_override
    
    if sym_seg == 0:
        return ""
    
    paa = piecewise_aggregate_approximation(pd.Series(normalized.values), sym_seg)
    cuts = get_cuts(paa, alphabet_size)
    symbolic_representation = cuts.apply(lambda cut: string.ascii_lowercase[cut - 1])
    return "".join(symbolic_representation)

interpolated_grouped_by_patient = pain_interpolated \
    .PainScoreQuantMissing \
    .groupby(level = "PatientID")
alpha = 3
saxes = interpolated_grouped_by_patient.apply(sax_unwindowed, alphabet_size=alpha)
```

```python
def get_unnormalized_cuts(paa_series, alphabet_size):
    alphabet_size = int(alphabet_size)
    #add really tiny amount to upper bound so that a pain score of 10 will still be considered less than that
    relevant_cut_points = np.linspace(0, 10.0001, alphabet_size + 1)
    relevant_cut_points_series = pd.Series(relevant_cut_points)
    cuts = paa_series.apply(lambda paa_element: np.sum(relevant_cut_points <= paa_element))
    return cuts
```

```python
def define_easy_categories(series, number_bins):
    cuts = get_unnormalized_cuts(series, number_bins)
    symbolic_representation = cuts.apply(lambda cut: string.ascii_lowercase[cut - 1])
    return "".join(symbolic_representation)

interpolated_grouped_by_patient = pain_interpolated \
    .PainScoreQuantMissing \
    .groupby(level = "PatientID")
k = 3
categories = interpolated_grouped_by_patient.apply(define_easy_categories, number_bins=k)
```

```python
all_motifs = {}

def get_raw_sequence_motif_vector(sequence, alphabet_size):
    l = 2 #motif length, hardcoded for reasons outlined in paper
    a = alphabet_size
    n = a ** l #possible motif combinations
    Mp = np.zeros(n) #motif vector
    for motif_pos in range(len(sequence) - l + 1):
        motif = sequence[motif_pos:motif_pos + l] #current motif
        i = 0
        for j in np.arange(l):
            kj = get_letter_index(motif[j])
            i += kj * a ** (l - j - 1)
        all_motifs[i] = motif
        Mp[i] += 1
    return pd.Series(Mp)

def get_letter_index(letter):
    return string.lowercase.index(letter)

motif_matrices_by_patient = saxes.apply(get_raw_sequence_motif_vector, alphabet_size=alpha)
```

```python
def get_clustered_motif_matrices(cluster_column):
    needed_columns = start_cutoff_pain_data[[cluster_column, "PatientID"]]
    unique_needed_columns = needed_columns.sort(columns=[cluster_column]).drop_duplicates()
    unique_needed_columns_indexed = unique_needed_columns.set_index(['PatientID'])
    cluster_info_joined_with_motifs = pd.concat([unique_needed_columns_indexed, motif_matrices_by_patient], axis=1)
    clustered_motrif_matrices = cluster_info_joined_with_motifs \
        .reset_index() \
        .dropna() \
        .set_index([cluster_column, "PatientID"]) \
        .sort_index()
    return clustered_motrif_matrices
```

```python
def sample_motif_importances(iteration, sample_size):
    sample_indices = np.random.choice(motif_matrices_by_patient.index, size=sample_size, replace=True)
    sample = motif_matrices_by_patient.ix[sample_indices]
    individual_motif_counts = sample.sum()
    total_motif_quantities = individual_motif_counts.sum(axis=1)
    x = individual_motif_counts / total_motif_quantities
    return x
```

```python
def M(cluster_sizes):
    n_X2_star = 1000
    X2_stars = []
    all_patients = patient_category_props.index
    n_clusters = len(cluster_sizes)
    categories = patient_category_props.iloc[0].index
    for i in range(n_X2_star):
        K_hat = pd.DataFrame(data=None, index = range(n_clusters), columns = categories)
        for j in range(n_clusters):
            category = categories[j]
            P_rand = np.random.choice(all_patients, size=cluster_sizes[j], replace=True)
            K_hat.ix[j] = patient_category_props.ix[P_rand].sum()
        #K_hat_exp = K_hat.sum(axis = 1).apply(lambda row_total: row_total * K_hat.sum(axis = 0)) / K_hat.sum().sum()
        #X2_star = ((K_hat_exp - K_hat) ** 2 / K_hat_exp).sum().sum()
        #this is faster
        X2_star = sc.stats.chi2_contingency(K_hat, correction=False)[0]
        X2_stars.append(X2_star)
    return X2_stars
```

```python
def get_expected_vals(cluster_sizes):
    n_X2_star = 1000
    X2_stars = []
    all_patients = patient_category_props.index
    n_clusters = len(cluster_sizes)
    categories = patient_category_props.iloc[0].index
    K_hat_sum = None
    for i in range(n_X2_star):
        K_hat = pd.DataFrame(data=None, index = range(n_clusters), columns = categories)
        for j in range(n_clusters):
            category = categories[j]
            P_rand = np.random.choice(all_patients, size=cluster_sizes[j], replace=True)
            K_hat.ix[j] = patient_category_props.ix[P_rand].sum()
        K_hat_exp = K_hat.sum(axis = 1).apply(lambda row_total: row_total * K_hat.sum(axis = 0)) / K_hat.sum().sum()
        if K_hat_sum is not None:
            K_hat_sum = K_hat_sum + K_hat_exp
        else:
            K_hat_sum = K_hat_exp
    return K_hat_sum / n_X2_star
```
