### interpolate the pain score values across consistent time intervals so that our SAX algorithm aligns accurately with real time

```python
def interpolate_patient_pain_scores(patient_dataframe_multiindexed, interval_minutes):
    #pandas automatically re-adds this in post-processing. remove it so we dont have a duplicate
    patient_dataframe_multiindexed.index = patient_dataframe_multiindexed.index.droplevel(0)
    if patient_dataframe_multiindexed.PainScoreQuantMissing.count() <= 1:
        return patient_dataframe_multiindexed
    patient_dataframe_multiindexed.dropna(subset=["PainScoreQuantMissing"], inplace=True)
    patient_dataframe_unindexed = patient_dataframe_multiindexed.reset_index()
    patient_dataframe_time_indexed = patient_dataframe_unindexed.set_index("NRSTimeFromEndSurgery_mins").sort_index()
    #use linear interpolation from scipy
    interpolation_func = interp1d(x=patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins, 
                                  y=patient_dataframe_unindexed.PainScoreQuantMissing)
    first_time_value = patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins.iloc[0]
    last_time_value = patient_dataframe_unindexed.NRSTimeFromEndSurgery_mins.iloc[-1]
    constant_time_interval_points = np.arange(first_time_value, last_time_value, interval_minutes)
    #expand the original data structure and add the interpolated values
    total_indexed_by_time = patient_dataframe_unindexed.set_index("NRSTimeFromEndSurgery_mins")
    total_indexed_by_time_expanded = total_indexed_by_time.reindex(constant_time_interval_points, method='ffill')
    total_indexed_by_time_expanded["PainScoreQuantMissing"] = [interpolated_val if interpolated_val > 0 else 0 for interpolated_val in interpolation_func(constant_time_interval_points)]
    return total_indexed_by_time_expanded

pain_data_indexed = start_cutoff_pain_data.set_index(["PatientID", "NRSTimeFromEndSurgery_mins"])
grouped_by_patient = pain_data_indexed.groupby(level=0)
pain_interpolated = grouped_by_patient.apply(interpolate_patient_pain_scores, 10)
pain_interpolated.dropna(subset=["PainScoreQuantMissing"], inplace=True)
#dont include patients with null time series after interpolation
patient_series_lengths = pain_interpolated.PainScoreQuantMissing.groupby(level=0).count()
patients_with_blank_series = patient_series_lengths[patient_series_lengths == 0]
pain_interpolated.drop(patients_with_blank_series.index, level=0, inplace=True)
interpolated_grouped_by_patient = pain_interpolated \
    .PainScoreQuantMissing \
    .groupby(level = "PatientID")
```

## SAX Calculation

```python
def piecewise_aggregate_approximation(series, paa_size):
    paa_size = int(paa_size)
    series_length = int(series.count())
    #take care of the special case where there is no dimensionality reduction
    if series_length == paa_size:
        paa = series
    else:
        if series_length % paa_size == 0:
            paa = series.reshape([series_length / paa_size, paa_size], order="F").mean(axis=0)
        else:
            temp = pd.DataFrame(series).T.reindex(index = xrange(paa_size), method="ffill")
            expanded_sub_section = temp.values.reshape(paa_size * series_length, order="F")
            paa = expanded_sub_section.reshape([series_length, paa_size], order="F").mean(axis=0)
    return pd.Series(paa)
```

```python
def get_unnormalized_cuts(paa_series, alphabet_size):
    alphabet_size = int(alphabet_size)
    #add really tiny amount to upper bound so that a pain score of 10 will still be considered less than that
    relevant_cut_points = np.linspace(0, 10.0001, alphabet_size + 1)
    relevant_cut_points_series = pd.Series(relevant_cut_points)
    cuts = paa_series.apply(lambda paa_element: np.sum(relevant_cut_points <= paa_element))
    return cuts
```

```python
def sax_unwindowed_unnormalized(series, alphabet_size, paa_length):
    paa = piecewise_aggregate_approximation(pd.Series(series.values), paa_length)
    cuts = get_unnormalized_cuts(paa, alphabet_size)
    symbolic_representation = cuts.apply(lambda cut: string.ascii_lowercase[cut - 1])
    return "".join(symbolic_representation)
```

### the following is a demonstration of the interpolation procedure (included here because uses PAA function)

```python
#    recorded=(&Minute\ 3: 6, Minute\ 24: 8, Minute\ 43: 9,\\
#              &Minute\ 100: 3, Minute\ 120: 1)
recorded={3:6, 24: 8, 55:9, 78:5, 115:4, 135: 1, 160: 2, 194: 1, 215: 6, 255: 9}
df = pd.DataFrame.from_dict(recorded, orient="index").sort_index()
df.columns = ["score"]
constant_time_interval_points = np.arange(np.min(recorded.keys()), np.max(recorded.keys()), 10)
points_to_fill = np.union1d(df.index, constant_time_interval_points)

f = interp1d(df.index, df["score"])

df_interpolated = df.reindex(index=constant_time_interval_points)
df_interpolated["score"] = f(constant_time_interval_points)
df_interpolated["score"] = df_interpolated["score"].apply(lambda score: float("%.1f" % score.round(1)))


paa_vals = piecewise_aggregate_approximation(pd.Series(df_interpolated.score.values),5)

fig = plt.figure(figsize=(7,4))
ax = fig.add_axes([0.15,0.1,.78,0.8])

plt.hold(True)
p1, = ax.plot(pd.Series(df.index) - df.index[0] , df["score"],'ko', figure=fig, label='Recorded')
p2, = ax.plot(constant_time_interval_points - df.index[0] , f(constant_time_interval_points), "kx", figure=fig, label='Interpolated')
p3, = ax.plot(points_to_fill - df.index[0] , f(points_to_fill), "k:", figure=fig, label='interpolation_line')
p4, = ax.plot([], [], "k-", figure=fig, label='paa_lines')
p5, = ax.plot([], [], "k--", figure=fig, label='sax_lines')
ax.legend([p1, p2], ["Recorded", "Interpolated"], loc='upper right', fontsize=12)
ax.set_ylim(0, 10)
ax.set_xlim(0, 260)
ax.set_title("Sample Pain Scores", fontsize=18, y=1.03)
ax.set_xlabel("$t$", fontsize=18)
ax.set_ylabel("$pain\ score$", fontsize=18)
fig.savefig(figure_save_dir + "interpolation.pdf", bbox_inches="tight")

show_paa=True
if show_paa:
    ax.hlines(paa_vals[0], 0, 50, linewidth=2)
    ax.hlines(paa_vals[1], 50, 100, linewidth=2)
    ax.hlines(paa_vals[2], 100, 150, linewidth=2)
    ax.hlines(paa_vals[3], 150, 200, linewidth=2)
    ax.hlines(paa_vals[4], 200, 250, linewidth=2)
    ax.legend([p1, p2, p4], ["Recorded", "Interpolated", "PAA"], loc='upper right', fontsize=12)
    fig.savefig(figure_save_dir + "paa.pdf", bbox_inches="tight")
    print("paa=" + str(paa_vals.values))
    
    
show_sax=True
if show_sax:
    ax.hlines(10./3., 0, 260, linewidth=1, linestyles="--")
    ax.hlines(2 * 10./3., 0, 260, linewidth=1, linestyles="--")
    ax.text(-0.17, 2 * 1./3. + (1./3. / 2) - 0.03, "c", transform=ax.transAxes, fontsize=24,
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'), zorder=100)
    ax.text(-0.17, 1 * 1./3. + (1./3. / 2) - 0.03, "b", transform=ax.transAxes, fontsize=24,
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'), zorder=100)
    ax.text(-0.17, (1./3. / 2) - 0.03, "a", transform=ax.transAxes, fontsize=24,
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'), zorder=100)
    #hack to avoid cliping the left margin
    ax.text(-0.2, (1./3. / 2) - 0.03, "a", transform=ax.transAxes, color="white", zorder=1)
    ax.legend([p1, p2, p4, p5], ["Recorded", "Interpolated", "PAA", "Breakpoint"], loc='upper center', fontsize=12)
    fig.savefig(figure_save_dir + "sax.pdf", bbox_inches="tight")
plt.hold(False)

print(", ".join(df_interpolated.reset_index().apply(lambda row: 'Minute ' + str(row["index"]) + ": " + str(row.score), axis=1).values))
print(",".join([str(val) for val in df_interpolated.reset_index().score.values]))
```

## Intelligent icon calculation

```python

def get_raw_sequence_motif_vector(sequence, alphabet_size):
    l = 2 #motif length, hardcoded for reasons outlined in paper
    a = alphabet_size
    n = a ** l #possible motif combinations
    Mp = np.zeros(n) #motif vector
    for motif_pos in range(len(sequence) - l + 1):
        motif = sequence[motif_pos:motif_pos + l] #current motif
        i = 0
        for j in np.arange(l):
            kj = get_letter_index(motif[j])
            i += kj * a ** (l - j - 1)
        all_motifs[i] = motif
        Mp[i] += 1
    return pd.Series(Mp)

def get_letter_index(letter):
    return string.lowercase.index(letter)
```

```python
def get_clustered_motif_matrices(cluster_column):
    needed_columns = start_cutoff_pain_data[[cluster_column, "PatientID"]]
    unique_needed_columns = needed_columns.sort(columns=[cluster_column]).drop_duplicates()
    unique_needed_columns_indexed = unique_needed_columns.set_index(['PatientID'])
    cluster_info_joined_with_motifs = pd.concat([unique_needed_columns_indexed, motif_matrices_by_patient], axis=1)
    clustered_motrif_matrices = cluster_info_joined_with_motifs \
        .reset_index() \
        .dropna() \
        .set_index([cluster_column, "PatientID"]) \
        .sort_index()
    return clustered_motrif_matrices
```

```python
def sample_motif_importances(iteration, sample_size):
    sample_indices = np.random.choice(motif_matrices_by_patient.index, size=sample_size, replace=True)
    sample = motif_matrices_by_patient.ix[sample_indices]
    individual_motif_counts = sample.sum()
    total_motif_quantities = individual_motif_counts.sum(axis=1)
    x = individual_motif_counts / total_motif_quantities
    return x
```

### Set everything up, including samples for histograms

```python
np.random.seed(10) #ensure we get consistent replicatable results
all_motifs = {}
beta = 10
number_of_ten_minute_intervals_in_frame = float(6 * 5) #5 hours
saxes = interpolated_grouped_by_patient.apply(lambda s: sax_unwindowed_unnormalized(s, alphabet_size=beta, paa_length = np.ceil(s.count() / number_of_ten_minute_intervals_in_frame)))
motif_matrices_by_patient = saxes.apply(get_raw_sequence_motif_vector, alphabet_size=beta)
nr = 1000
example_ck = 100
motif_importance_sample = pd.Series(np.arange(nr)).apply(sample_motif_importances, sample_size=example_ck)
```

```python
#start_cutoff_pain_data.PrimaryCPTCodeCategory2
#clustered_motif_matrices = get_clustered_motif_matrices("PrimaryCPTCodeCategory2")
#clustered_motif_matrices = get_clustered_motif_matrices("Gender")
clustered_motif_matrices = get_clustered_motif_matrices("custom_column")
#clustered_motif_matrices = get_clustered_motif_matrices("Age_Group")
clustered_motif_matrices_grouped = clustered_motif_matrices.groupby(level=0)
cluster_sample_sizes = clustered_motif_matrices_grouped.count()
individual_motif_counts_by_cluster = clustered_motif_matrices_grouped.sum()
total_motif_quantities_by_cluster = individual_motif_counts_by_cluster.sum(axis=1)
motif_proportions_clusters = individual_motif_counts_by_cluster.apply(lambda val: val / total_motif_quantities_by_cluster)
cluster_criteria = motif_proportions_clusters.index
icons = pd.DataFrame(columns = motif_matrices_by_patient.columns, index = pd.Index(data=[], name="theta"))
for criterion in cluster_criteria:
    motif_proportions_cluster = motif_proportions_clusters.ix[criterion]
    cluster_n = cluster_sample_sizes.ix[criterion][0]
    minimum_cluster_n = 50
    if cluster_n > minimum_cluster_n:
        print("Generating Xci for theta=" + str(criterion) + " - nc=" + str(cluster_n))
        nr = 1000
        motif_importance_sample = pd.Series(np.arange(nr)).apply(sample_motif_importances, sample_size=cluster_n)
        numerators = motif_importance_sample.apply(lambda row: row <= motif_proportions_cluster, axis=1).sum()
        denominators = motif_importance_sample.count()
        CDF = numerators / denominators
        icon = CDF
        icons.ix[str(criterion)] = icon
    else:
        print("Skipping theta=" + str(criterion) + " - nc=" + str(cluster_n))
    
icons
```

```python
def get_similarities(icon, all_icons):
    similarities = cosine_similarity(X=icon, Y=all_icons)[0]
    return pd.DataFrame(data=similarities, 
                        index=pd.Index(all_icons.index, name="other_item"),
                        columns=pd.Index(["similarity"]))

all_theta_similarities = icons.groupby(level=0).apply(get_similarities, all_icons=icons)
all_theta_similarities
```
